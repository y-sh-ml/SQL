```python
import config
import psycopg2
import pandas as pd
```

```python
conn = psycopg2.connect(**config.connection_config)
cursor = conn.cursor()
```

```python
query ="""

SELECT
  *
FROM
  customer_tb
LIMIT
  5

"""

pd.read_sql(query, conn)
```

```python
query ="""

CREATE TABLE
  customer_m40over
AS

SELECT
  customer_id,
  age,
  sex
FROM
  customer_tb
WHERE
  age >= 40
  AND sex = 'man'

"""

cursor.execute(query)
conn.commit()
```

```python
query ="""

SELECT
  *
FROM
  customer_m40over
LIMIT
  5

"""

pd.read_sql(query, conn)
```

```python
query ="""

DROP TABLE
  customer_m40over

"""

cursor.execute(query)
conn.commit()
```
